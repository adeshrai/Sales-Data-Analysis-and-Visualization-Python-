```python
import pandas as pd
data = pd.read_csv('sales_data.csv')
print(data.isnull().sum())
data['Customer_Age'] = data['Customer_Age'].fillna(data['Customer_Age'].median())

```

```python
# Total sales by product
product_sales = data.groupby('Product')['Total_Sale'].sum().sort_values(ascending=False)

# Monthly sales trend
data['Order_Date'] = pd.to_datetime(data['Order_Date'], dayfirst=True)

# Extract month for monthly trend analysis
data['Month'] = data['Order_Date'].dt.month

# Calculate monthly sales
monthly_sales = data.groupby('Month')['Total_Sale'].sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# To Group city sales
city_sales = data.groupby('City')['Total_Sale'].sum().sort_values(ascending=False)

# TobCreate age group segmentation
data['Age_Group'] = pd.cut(data['Customer_Age'], bins=[0, 18, 35, 50, 100], labels=['Teen', 'Young Adult', 'Adult', 'Senior'])
age_group_sales = data.groupby('Age_Group', observed=True)['Total_Sale'].sum()

# TO Set up a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# For Top-selling products (Bar chart)
product_sales.head(10).plot(kind='bar', color='skyblue', ax=axs[0, 0])
axs[0, 0].set_title('Top-Selling Products')
axs[0, 0].set_ylabel('Total Sales')

# For Monthly sales trend (Line plot)
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, ax=axs[0, 1])
axs[0, 1].set_title('Monthly Sales Trend')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Total Sales')

# For Top cities by sales (Horizontal bar chart)
city_sales.head(10).plot(kind='barh', color='orange', ax=axs[1, 0])
axs[1, 0].set_title('Top Cities by Sales')
axs[1, 0].set_xlabel('Total Sales')
axs[1, 0].set_ylabel('City')

# Sales by age group (Pie chart)
age_group_sales.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'lightgreen', 'lightblue', 'pink'],
    ax=axs[1, 1]
)
axs[1, 1].set_title('Sales by Age Group')
axs[1, 1].set_ylabel('')

plt.tight_layout()
plt.show()
```
